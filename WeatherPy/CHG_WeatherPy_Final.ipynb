## Three Observable Trends

1. Temperatures increase as you move towards the equator
2. Wind Speed does not seem to be predicted by Latitude
3. Cloudiness does not seem to be predicted by Latitude

```python
# bring in dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
```

```python
# bring in api key
from api_keys import weather_api_key
```

```python
# Import citiby to grab cities by their lat/long numbers
from citipy import citipy
```

```python
# output file
output = "cities.csv"
```

```python
# range of latitudes and longitude
lat_range = (-90,90)
long_range = (-180,180)
```

```python
# list for holding coordinates and citie names
coordinates = []
city_names = []

# create random set of coordinates with 1500 entries
# np.random.uniform creates a sample of numbers in which any value
# within the given interval is equally likely to be drawn

lat = np.random.uniform(lat_range[0], lat_range[1], size=1500)
long = np.random.uniform(long_range[0], long_range[1], size=1500)
coordinates = zip(lat,long)
```

```python
# identify nearest city for each coordinate
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # if the city is unique, then add it to our cities list
    if city not in city_names:
        city_names.append(city)
```

```python
# print the city count to confirm we have the sample number we coded in
len(city_names)
```

## Generating My CSV File With All Needed Variables

For my CSV, I have grabbed the latitude, longitude, name, temperature in Farenheit, humidity percentage, cloudiness percentage,
wind speed, country, and date

```python
# running an api request on the url to get the temperatures of each city in the city_names list
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"

# Lists to hold city information

lat2 = []
long2 = []
cities = []
temperature = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []
items = []
count = 0
# Build query URL
print('Beginning Data Retrieval')
print('----------------------------')
print('')
print('')

for city in city_names:
    query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=imperial"
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    try:
        lat2.append(weather_json['coord']['lat'])
        items.append(1)
        long2.append(weather_json['coord']['lon'])
        items.append(2)
        cities.append(weather_json['name'])
        items.append(3)
        temperature.append(weather_json['main']['temp'])
        items.append(4)
        humidity.append(weather_json['main']['humidity'])
        items.append(5)
        cloudiness.append(weather_json['clouds']['all'])
        items.append(6)
        wind_speed.append(weather_json['wind']['speed'])
        items.append(7)
        country.append(weather_json['sys']['country'])
        items.append(8)
        date.append(weather_json['dt'])
        items.append(9)
        count+= 1
        print(f"Processing Data for City # {count} | Name: {city}... ")
        print('------------------------------------------------------')
        print(f"    Collected Stat #{items[0]} of {len(items)}..." )
        print(f"    Collected Stat #{items[1]} of {len(items)}..." )
        print(f"    Collected Stat #{items[2]} of {len(items)}..." )
        print(f"    Collected Stat #{items[3]} of {len(items)}..." )
        print(f"    Collected Stat #{items[4]} of {len(items)}..." )
        print(f"    Collected Stat #{items[5]} of {len(items)}..." )
        print(f"    Collected Stat #{items[6]} of {len(items)}..." )
        print(f"    Collected Stat #{items[7]} of {len(items)}..." )
        print(f"    Collected Stat #{items[8]} of {len(items)}..." )
        print('------------------------------------------------------')
        print('------------------------------------------------------')
        print(f"Data Processing for {city} COMPLETE")
        print('------------------------------------------------------')
        print('------------------------------------------------------')
        print('')
        print('Moving to Next City...')
        print('')
        print('')
        items = []
    except:
        print('------------------------------------------------------')
        print('------------------------------------------------------')
        print(f"The requested city ...'{city}'... was not found via the API Request")
        print('------------------------------------------------------')
        print('------------------------------------------------------')
        print('')
        print('Moving to Next City...')
        print('')
        print('')
        pass
```

```python
# zip the collected data into a dataframe to be exported

weather_data = {"City Name":cities,
               "Latitude":lat2,
                "Longitude":long2,
                "Temperature (Farenheit)":temperature,
                "Humidity %":humidity,
                "Cloudiness %":cloudiness,
                "Wind Speed (MPH)":wind_speed,
               "Country":country,
               "Date":date}

weather_df = pd.DataFrame(weather_data)
```

```python
# export dataframe to csv files
weather_df.to_csv('cleaned_weather_data.csv')
```

```python
# calculate to see if we have any humidity outliers

for x in weather_df['Humidity %']:
    if x > 100:
        print('Found Outlier')
```

```python
# creating my northern and southern hemisphere dataframes

northern = weather_df[weather_df['Latitude'] >= 0]

southern = weather_df[weather_df['Latitude'] <= 0]
```

```python
# creating a scatterplot function

def create_scatter(x,y,c):
    (slope, intercept, rvalue, pvalue, stderr) = st.linregress(x, y)
    regress_values = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x,y,color=c)
    plt.plot(x,regress_values,"r-")
    plt.annotate(line_eq,(2,2),fontsize=15,color="green")
    plt.xlabel('Latitudes')
```

```python
# creating variables for scatter plot creation

n_temp = northern['Temperature (Farenheit)']
s_temp = southern['Temperature (Farenheit)']
n_hum = northern['Humidity %']
s_hum = southern['Humidity %']
n_clo = northern['Cloudiness %']
s_clo = southern['Cloudiness %']
n_wind = northern['Wind Speed (MPH)']
s_wind = southern['Wind Speed (MPH)']

n_lat = northern['Latitude']
s_lat = southern['Latitude']
```

```python
# Northern Temp vs Latitude

create_scatter(n_lat,n_temp,'green')
plt.title('Northern Hemisphere Cities Temperature (F) Against Latitude')
plt.ylabel('Temperature (F)')
plt.savefig('Charts/north_temp.png')
plt.show()
```

```python
# Southern Temp vs Latitude

create_scatter(s_lat,s_temp,'purple')
plt.title('Southern Hemisphere Cities Temperature (F) Against Latitude')
plt.ylabel('Temperature (F)')
plt.savefig('Charts/south_temp.png')
plt.show()
```

### Analysis

After modeling the northern and southern hemispheres temperature against the latitudes, it is clear that northern hemisphere
cities get cooler as their latitude gets greater and we see the opposite in the southern hemisphere cities

```python
# Northern Humidity vs Latitude

create_scatter(n_lat,n_hum,'yellow')
plt.title('Northern Hemisphere Cities Humidity Percentage Against Latitude')
plt.ylabel('Humidity %')
plt.savefig('Charts/north_hum.png')
plt.show()
```

```python
# Southern Humidity vs Latitude

create_scatter(s_lat,s_hum,'blue')
plt.title('Southern Hemisphere Cities Humidity Percentage Against Latitude')
plt.ylabel('Humidity %')
plt.savefig('Charts/south_hum.png')
plt.show()
```

### Analysis

After modeling the northern and southern hemispheres humidity percentages across lattitudes, it does not appear that humidity and latitude have much of a relationship 

```python
# Northern Cloudiness vs Latitude

create_scatter(n_lat,n_clo,'pink')
plt.title('Northern Hemisphere Cities Cloudiness Percentage Against Latitude')
plt.ylabel('Cloudiness %')
plt.savefig('Charts/north_clo.png')
plt.show()
```

```python
# Southern Cloudiness vs Latitude

create_scatter(s_lat,s_clo,'orange')
plt.title('Southern Hemisphere Cities Cloudiness Percentage Against Latitude')
plt.ylabel('Cloudiness %')
plt.savefig('Charts/south_clo.png')
plt.show()
```

### Analysis

After modeling the northern and southern hemispheres cloudiness percentages - it appears that there is no relationship in the northern cities, but when you look at the southern cities, you can see that cloudiness decreases as you move closer to 0 in latitude
